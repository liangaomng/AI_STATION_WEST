```python
import torch
import matplotlib.pyplot as plt
```

```python
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

class CustomDataset(Dataset):
    def __init__(self, file_path):
        # 读取.pt文件
        data = torch.load(file_path)
        self.data = data['data']
        self.label = data['label_csv']
        self.length = len(self.data)
    
    def __getitem__(self, index):
        # 获取数据和标签
        data = self.data[index]
        label = self.label[index]
        
        # 数据预处理操作
        
        return data, label
    
    def __len__(self):
        return self.length
train_loader = torch.load("2nd_dataset/2ndtrain.pth")
test_loader = torch.load("2nd_dataset/2ndtest.pth")
```

```python
# to show batch_data
for i, batch in enumerate(test_loader):
    batch_data=batch[0]
    t=batch_data[0,:,6]
    z1_t=batch_data[0,:,7]
    z2_t=batch_data[0,:,8]
    u=[x - y for x, y in zip(z1_t[1:], z1_t[:-1])]
    v=[x-y for x,y in zip(z2_t[1:],z2_t[:-1])]
    #phase 
    #quiver
    plt.style.use('ggplot')
    plt.plot(z1_t,z2_t,color='b')
    plt.quiver(z1_t[:-1], z2_t[:-1],
               u,
               v,scale_units='xy', angles='xy', scale=0.3,color='r')
    plt.xlabel("z1(t)")
    plt.ylabel("z2(t)")

plt.show()   
```
